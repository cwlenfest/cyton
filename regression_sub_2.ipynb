```python
import statistics as stat
import keras
import sklearn
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn import preprocessing
from sklearn import model_selection


mse=np.zeros(50)
for i in range(50):
    model = Sequential()
    dataset = pd.read_csv("Concrete_Data.csv")
    dataset = preprocessing.normalize(dataset)
    
    training_data, test_data = model_selection.train_test_split(dataset, test_size=0.3)
    
    x_train = training_data[:, 0:8]
    x_test = test_data[:, 0:8]
    y_train = training_data[:, 8]
    y_test = test_data[:, 8] 
    n_cols = training_data.shape[1]
    model.add(Dense(10,activation='relu',input_shape=(n_cols-1,)))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(x_train, y_train, epochs=50);

    y_predict = model.predict(x_test)

    mse[i]=sklearn.metrics.mean_squared_error(y_test,y_predict)

mean_mse=stat.mean(mse) #Find the mean of the mean squared error
sd_mse=stat.stdev(mse) #Find the standard deviation of the mean squared error
print("Mean",mean_mse)
print("SD",sd_mse)

```
